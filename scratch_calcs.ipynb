```python
from pyscf import gto, scf
```

```python
mol = gto.M(
    atom = '''
O        0.000000    0.000000    0.117790
H        0.000000    0.755453   -0.471161
H        0.000000   -0.755453   -0.471161''',
    basis = 'ccpvdz',
    charge = 1,
    spin = 1  # = 2S = spin_up - spin_down
)
```

```python
mf = scf.UHF(mol)
mf.kernel()
```

```python
print(mf.mo_coeff.shape)
print(mol.nelectron)
```
