```python
from pyscf import gto, scf
```

```python
mol = gto.M(
    atom = '''
O        0.000000    0.000000    0.117790
H        0.000000    0.755453   -0.471161
H        0.000000   -0.755453   -0.471161''',
    basis = 'sto-3g',
    charge = 1,
    spin = 1  # = 2S = spin_up - spin_down
)
```

```python
mf = scf.UHF(mol)
mf.kernel()
```

```python
mf.scf_summary
```

```python
nelec = mol.nelectron
nao = mol.nao_nr()
mo = mf.mo_coeff
norb = mo.shape[1]
hcore = mf.get_hcore()
eri_ao = mf._eri
print(nelec)
```

```python
from hci import hci_uhf
hci_uhf.kernel(hcore, eri_ao, mo, norb, (5, 4))
```
